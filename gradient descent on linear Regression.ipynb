```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
x_train = [2,3,4,5,7,8,10,12,13,16,18,21,24,27,32,34,37]
y_train = [100,150,130,180,200,230,300,500,190,700,430,450,800,900,1000,950,1100]
```

```python
plt.scatter(x_train,y_train,c='r',marker = 's',edgecolors = 'w',s=50)
plt.title('salary prediction')
plt.xlabel('experience')
plt.ylabel('salary')
plt.show()
```

```python
def linearRegression(x_data,teta1,teta2):
    m = len(x_data)
    prd = np.zeros(m)
    for i in range(m):
        prd[i]=(teta2*x_data[i]) + teta1
    return prd
        
```

```python
w = 20
b = 10
Reg = linearRegression(x_train,b,w)
```

```python
plt.scatter(x_train,y_train,c='r',marker = 's',edgecolors = 'w',s=50)
plt.plot(x_train,Reg,c = 'g')
plt.title('salary prediction')
plt.xlabel('experience')
plt.ylabel('salary')
plt.show()
```

```python

def regression(x_data,teta1,teta2):
    m = len(x_data)
    prd = np.zeros(m)
    for i in range(m):
        prd[i]=(teta2*x_data[i]) + teta1
    return prd
def cost(prediction,actual):
    m = len(prediction)
    cost = 0
    for i in range(m):
        diff = (prediction[i]-actual[i])**2
        cost = cost + diff
    cost = (1/(2*m))*cost
    return cost
def gradient(x_data,prediction,actual):
    m = len(x_data)
    w_it = 0
    b_it =0
    for i in range(m):
        wigt = ( prediction[i]-actual[i]) * x_data[i]
        bais = prediction[i]- actual[i]
        w_it +=wigt
        b_it+= bais
    w_it = w_it/m
    b_it = b_it/m
    return w_it,b_it
def gradient_descent(x_data,actual,lr,itr):
    m = len(x_data)
    cost_fn = []
    w_b =[] 
    w,b = 50,70
    for i in range(itr):
        prd = regression(x_data,b,w)
        w_i,b_i =gradient(x_data,prd,actual)
        w = w-lr*(w_i)
        b = b - lr*(b_i)
        cost_fn.append(cost(prd,actual))
        w_b.append([w,b])
        
    return w,b,prd,w_b,cost_fn
```

```python
widet,bais,pred,w_and_b,cost = gradient_descent(x_train,y_train,0.001,10000)
```

```python
plt.scatter(x_train,y_train,c='r',marker = 's',edgecolors = 'w',s=50)
plt.plot(x_train,pred,c = 'g')
plt.title('salary prediction')
plt.xlabel('experience')
plt.ylabel('salary')
plt.show()
```
