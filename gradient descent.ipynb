```python
# initializing the parameter
w = 250
b = 200
print(f'parameter w : {w}, b : {b}')
```

```python
# prediction of model
def linearRegression(x_data,theta1,theta2):
    
# m is total no of train example
    m = len(x_data)
    predict = np.zeros(m)
    for i in range(m):
        predict[i]=theta1 + (theta2*x_data[i])
        
    return predict
```

```python
Regerssion = linearRegression(x_train,b,w)
```

```python
# ploting the data
plt.scatter(x_train,y_train,c='r',s=50,marker='s',edgecolors='w',label = 'training data')
plt.plot(x_train,Regerssion,c='g',label ='prediction')
plt.xlabel("size of house in 1000 sqr foot")
plt.ylabel("price of house in 1000 dollar")
plt.title('housing price')
plt.show()
```

```python
# compute gradient
def gradient_descent(x_data,predit,actual):
    m = len(predit)
    gd_w = 0
    gd_b =0
    for i in range(m):
        reg = (predit[i]-actual[i]) * x_data[i]
        bais = predit[i]-actual[i]
        gd_w = gd_w + reg
        gd_b = gd_b + bais
    gd_w = gd_w/m
    gd_b = gd_b/m
    return gd_w,gd_b
     

    
```

calculate the gradient

```python
def calculate_gradient(x_data,theta1,theta2,actual,iteration,lr):
    m = len(x_data)
    w_it = theta2
    b_it = theta1
    w_hit = []
    b_hit= []
    for i in range(iteration):
        prd =  linearRegression(x_data,b_it,w_it) 
        w_i,b_i = gradient_descent(x_data,prd,actual)
        w_it = w_it- (lr * w_i)
        b_it = b_it - (lr * b_i)
        w_hit.append(cost_function(prd,y_train))
        b_hit.append([w_it,b_it])
        if i % 5000 == 0:
              print(f'w  : {int(w_it)},b : {int(b_it)}')
    return w_it,b_it,w_hit,b_hit,prd
         
          
```
